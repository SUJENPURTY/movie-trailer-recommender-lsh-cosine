## Why `dropna()` Instead of `fillna()`?
For recommendation tags, fields like `overview`, `genres`, `cast`, and `crew` carry core semantic meaning. Filling missing values with placeholders (such as empty strings or defaults) can add noisy or misleading tokens and weaken similarity quality. Dropping incomplete rows preserves cleaner, higher-signal feature text for modeling.

### 9) Parse JSON-like columns using ast.literal_eval

```python
def extract_name_list(text_value, limit=None):
    parsed = ast.literal_eval(text_value)
    names = [item["name"] for item in parsed if isinstance(item, dict) and "name" in item]
    if limit is not None:
        names = names[:limit]
    return names


def extract_director(text_value):
    parsed = ast.literal_eval(text_value)
    directors = [
        item["name"]
        for item in parsed
        if isinstance(item, dict) and item.get("job") == "Director" and "name" in item
    ]
    return directors[:1]
```

```python
movies_df["genres"] = movies_df["genres"].apply(extract_name_list)
movies_df["keywords"] = movies_df["keywords"].apply(extract_name_list)
movies_df["cast"] = movies_df["cast"].apply(lambda x: extract_name_list(x, limit=3))
movies_df["crew"] = movies_df["crew"].apply(extract_director)
```

```python
print("Parsed sample:")
print(movies_df[["title", "genres", "keywords", "cast", "crew"]].head(2))
```

```python
# 10) Process overview text and normalize names
movies_df["overview"] = movies_df["overview"].apply(lambda x: x.split())


def remove_inner_spaces(token_list):
    return [token.replace(" ", "") for token in token_list]

for feature_col in ["genres", "keywords", "cast", "crew"]:
    movies_df[feature_col] = movies_df[feature_col].apply(remove_inner_spaces)

print("Processed text sample:")
print(movies_df[["title", "overview", "cast", "crew"]].head(1))
```

## Create Feature-Rich `tags` Column
Combine text features (`overview`, `genres`, `keywords`, `cast`, `crew`) into one modeling column.

### 11) Create tags by combining all selected feature lists

```python
movies_df["tags"] = (
    movies_df["overview"]
    + movies_df["genres"]
    + movies_df["keywords"]
    + movies_df["cast"]
    + movies_df["crew"]
)
```

```python
print("Tags list sample:")
print(movies_df[["title", "tags"]].head(2))
```

### 12) Final cleaning: convert tags list to lowercase text

```python
movies_df["tags"] = movies_df["tags"].apply(lambda values: " ".join(values).lower())

final_movies_df = movies_df[["movie_id", "title", "tags"]].copy()
```
